**a.ravel(): (i) Return only reference/view of original array (ii) If you modify the array you would notice that the value of original array also changes. (iii) Ravel is faster than flatten() as it does not occupy any memory. (iv) Ravel is a library-level function.
a.flatten(): (i) Return copy of original array (ii) If you modify any value of this array value of original array is not affected. (iii) Flatten() is comparatively slower than ravel() as it occupies memory. (iv) Flatten is a method of an ndarray object.**

plotting trignometric curves

```python
x_sin = np.arange(0,2*np.pi,0.1)
y_sin = np.sin(x_sin)
print(y_sin)


```

```python
plt.figure(figsize=(6,6))
plt.plot(x_sin,y_sin,)
plt.title('sine wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x_cos = np.arange(0,2*np.pi,0.1)
y_cos = np.cos(x_cos)
print(y_cos)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x_cos,y_cos)
plt.title('cos wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x_tan = np.arange(0,2*np.pi,0.1)
y_tan = np.tan(x_tan)
print(y_tan)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x_tan,y_tan,)
plt.title('tan wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x_cot = np.arange(0,2*np.pi,0.1)
y_cot = 1/np.tan(x_cot)
print(y_cot)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x_cot,y_cot,)
plt.title('cot wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

subplot

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.plot(x_sin,y_sin,'r-')
plt.title('sine wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(2,2,2)
plt.plot(x_cos,y_cos,'b-')
plt.title('cosine wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(2,2,3)
plt.plot(x_tan,y_tan,'g-')
plt.title('tan wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(2,2,4)
plt.plot(x_cot,y_cot,'w-')
plt.title('cot wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```
