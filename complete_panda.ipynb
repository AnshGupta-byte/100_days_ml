# Merge() Function

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Physics':[34,67,34,89,12]})
df1
```

```python
df2 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Chemistry':[78,33,39,81,90]})
df2
```

```python
pd.merge(df1, df2, on = 'Roll No.')
```

```python
pd.merge(df2, df1, on = 'Roll No.')
```

```python
pd.merge(df1, df2)   # on - by default intersecting column is taken
```

```python
df3 = pd.DataFrame({'Roll No.':[1,2,3,6,7], 'Physics':[34,67,34,89,12]})
df4 = pd.DataFrame({'Roll No.':[1,2,3,8,9], 'Chemistry':[34,67,34,89,12]})
```

```python
pd.merge(df3, df4)
```

```python
pd.merge(df3, df4, how = 'left')
```

```python
pd.merge(df3, df4, how = 'right')
```

```python
pd.merge(df3, df4, how = 'outer')
```

# append() Function

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Maths':[45,78,45,90,66], 'Physics':[33,67,12,90,44]})
df2 = pd.DataFrame({'Roll No.':[6,7,8,9,10], 'Maths':[78,73,45,90,69], 'Physics':[23,67,88,0,98]})
print(df1)
print(df2)
```

```python
df = pd.concat([df1, df2]) #  .append is decaprecated from panda so we use .concat
df
```

```python
df = pd.concat([df2, df1]) #  .append is decaprecated from panda so we use .concat
df
```

```python
df = pd.concat([df1, df2],ignore_index=True) #  .append is decaprecated from panda so we use .concat
df
```

```python
df = pd.concat([df1, df2],ignore_index=True,sort=True) #  .append is decaprecated from panda so we use .concat
df
```

# Pivot Table

```python
df = pd.read_csv('/content/sample2.csv')
df.head()
```

```python
pd.pivot_table(df, index='Branch',
               values=['Chemistry','Computer','Maths','Physics'])
```

```python
pd.pivot_table(df, index='Branch',values=['Chemistry','Computer','Maths','Physics'], aggfunc='sum')
```

```python
pd.pivot_table(df, index='Branch', aggfunc='count')
```

```python
pd.pivot_table(df, index='Branch', aggfunc='max')
```

```python
pd.pivot_table(
    df,
    index='Branch',
    columns='Section',
    values=['Chemistry', 'Computer', 'Maths', 'Physics'],
    aggfunc='mean'
)
```
